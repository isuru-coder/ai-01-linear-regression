```python
# --- Day 4: Cross-validation + Pipeline (beginner-friendly) ---

import json
import os
import numpy as np
import pandas as pd

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold, cross_validate
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

# 1) Load data
df = pd.read_csv(DATA_PATH)
assert {"years_experience", "salary"}.issubset(df.columns), "CSV must have years_experience,salary"

X = df[["years_experience"]].values   # feature matrix (2D)
y = df["salary"].values               # target (1D)

# 2) Build a pipeline: [Standardize] -> [Linear Regression]
pipe = Pipeline(steps=[
    ("scaler", StandardScaler()),
    ("model", LinearRegression())
])

# 3) Define cross-validation: 5 folds, shuffle for fairness, fixed seed for reproducibility
cv = KFold(n_splits=5, shuffle=True, random_state=42)

# 4) Choose metrics
# - R^2: higher is better
# - MAE: lower is better (scikit-learn returns negative MAE; we'll flip the sign later)
scoring = {
    "r2": "r2",
    "mae": "neg_mean_absolute_error"
}

# 5) Run cross-validation for both metrics at once
cv_results = cross_validate(
    estimator=pipe,
    X=X, y=y,
    cv=cv,
    scoring=scoring,
    return_train_score=False
)

# 6) Extract per-fold metrics
r2_folds = cv_results["test_r2"].tolist()
mae_folds_neg = cv_results["test_mae"].tolist()     # negative numbers
mae_folds = (-np.array(mae_folds_neg)).tolist()     # make them positive

# 7) Compute summary (mean/std)
r2_mean, r2_std = float(np.mean(r2_folds)), float(np.std(r2_folds, ddof=1))
mae_mean, mae_std = float(np.mean(mae_folds)), float(np.std(mae_folds, ddof=1))

# 8) Prepare JSON payload
payload = {
    "model": "Pipeline(StandardScaler -> LinearRegression)",
    "cv": {"n_splits": 5, "shuffle": True, "random_state": 42},
    "metrics": {
        "r2":   {"folds": r2_folds,  "mean": r2_mean,  "std": r2_std},
        "mae":  {"folds": mae_folds, "mean": mae_mean, "std": mae_std}
    }
}

# 9) Save to reports/metrics_cv.json (create folder if missing)
os.makedirs("reports", exist_ok=True)
with open("reports/metrics_cv.json", "w", encoding="utf-8") as f:
    json.dump(payload, f, indent=2)

# 10) Print a friendly summary
print("Cross-validation complete!")
print(f"R^2 folds : {np.round(r2_folds, 4)} | mean={r2_mean:.4f}  {r2_std:.4f}")
print(f"MAE folds : {np.round(mae_folds, 2)}    | mean={mae_mean:.2f}  {mae_std:.2f}")
print('Saved -> reports/metrics_cv.json')
```
